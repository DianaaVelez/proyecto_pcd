```python
# Contar publicaciones positivas dentro del rango de horas laborales para cada pa√≠s
def count_positive_work(sentiments, rango_paises):
    resultados = []

    for pais, rangos in rango_paises.items():
        # Obtener rango de trabajo
        work_start, work_end = rangos['work']
        
        # Filtrar publicaciones dentro del rango laboral y con sentimiento positivo
        datos_work_positive = sentiments[
            (sentiments['Hour'] >= work_start) &
            (sentiments['Hour'] <= work_end) &
            (sentiments['Sentiment'] == 'Positive')
        ]
        
        # Contar publicaciones positivas
        total_positivas = len(datos_work_positive)
        
        # Agregar resultado a la lista
        resultados.append({'Pa√≠s': pais, 'Publicaciones Positivas': total_positivas})
    
    # Convertir los resultados a un DataFrame
    return pd.DataFrame(resultados)

# Aplicar la funci√≥n
resultados_positivos = count_positive_work(sentiments, rango_paises)
resultados_positivos
```

```python
def count_positive_work(sentiments, rango_paises):
    resultados = []

    for pais, rangos in rango_paises.items():
        # Obtener rango de trabajo
        work_start, work_end = rangos['work']
        
        # Filtrar publicaciones del pa√≠s actual
        datos_pais = sentiments[sentiments['Country'] == pais]
        
        # Filtrar publicaciones dentro del rango laboral y con sentimiento positivo
        datos_work_positive = datos_pais[
            (datos_pais['Hour'] >= work_start) &
            (datos_pais['Hour'] <= work_end) &
            (datos_pais['Sentiment'] == 'Positive')
        ]
        
        # Contar publicaciones positivas
        total_positivas = len(datos_work_positive)
        
        # Agregar resultado a la lista
        resultados.append({'Pa√≠s': pais, 'Publicaciones Positivas': total_positivas})
    
    # Convertir los resultados a un DataFrame
    return pd.DataFrame(resultados)

# Aplicar la funci√≥n
resultados_positivos = count_positive_work(sentiments, rango_paises)
resultados_positivos 
```
